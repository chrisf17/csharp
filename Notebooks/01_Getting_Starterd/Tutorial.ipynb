# C# Getting started

## What is C# #

C# is an open source, object oriented programming language. 

### Key features

* Works with .NET Libraries (.NET Core or Framework)
* Case sensitive
* Whitespace agnostic

### Simple C# Program

The simplest C# programs are called console applications. Console apps read and write to the console window using a .NET Library class called **Console**.

The Console class writes to the console window like this
```
Console.WriteLine("Hello World");
```

Notice how it has a semicolon *;* at the end of the statement. All C# statements are terminated with ;

#### Hands On

````
using System;
Console.WriteLine("Hello World");
````

### Comments

C# allows you to add comments to your code that are ignored by the compiler.

Two forward slashes // act as a line comment 

````
// We will now definea person's age
int age = 21; // person's age
````

Everything between /* and */ is also commented out accross multiple line.



````
int age = 21;
/*

This is all commented out
Console.WriteLine("I do nothing");

*/

Console.WriteLine(age);
````

## Variables

Variables allow us to store data in memory so that we can use that data later in our program.

### Declaring Variables

In C# variables have the following syntax.

```
<variable-type> <variable-name> ;

```

An example would be:

```

string name = "Beth";

```

### Naming Variables

Variable names must start with alpha characters or underscore ( _ ). They cannot start with number characters or any other punctuation or operator characters.

```
int age;   // OK
int _age;  // OK
int 5Lives // Not OK!! Cannot start with 5
```
#### Hands On
* Run the code below
* Now run it after uncommenting the commented out line

````
int _age = 21;
//int 5Lives = 99;
Console.WriteLine("All is Well");
````

### Initialising Variables

You can initialise a variable when you declare it.  

```

int age = 21;

```

Or you can do it later.

```
int age;


age = 21;
```

**Look at the following example**

````
int age;

age = 21;

Console.WriteLine(age);
````

````
int age = 22;

Console.WriteLine(age);
````

#### Rule: Initialize variables before you use them
